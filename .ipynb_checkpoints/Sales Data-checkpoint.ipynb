```python
charger = dodge[dodge['month'] != 'Or']
```

```python
charger['month'] = charger['month'].astype('int32')
```

### What are the products which are sold together.

```python
charger.head()
```

```python
charger['Order ID'].nunique()
```

```python
charger.groupby('Order ID')['Product'].count()
```

```python
dfg = charger[charger['Order ID'].duplicated(keep = False)]
dfg
```

```python
dfg['grouped'] = dfg.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
dfg.head()
```

```python
asd = dfg.drop_duplicates(subset = 'Order ID')
```

```python
asd.head()
```

### Which products have been sold mostly pairwise.

```python
rick = asd[['Order ID', 'grouped']].reset_index(drop = True)
```

```python
rick
```

```python
rick.groupby('grouped').count().sort_values(by = 'Order ID', ascending = False)
```

```python
count = Counter()
for i in rick['grouped']:
    rowlist = i.split(',')
    count.update(Counter(combinations(rowlist, 2)))
print(count)
```

```python
for key, value in count.most_common(10):
    print(key, value)
```
